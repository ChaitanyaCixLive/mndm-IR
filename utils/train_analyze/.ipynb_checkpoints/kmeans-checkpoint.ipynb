```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
```

```python
def get_name2feat():
    names = np.load('train_filename.npy').tolist()
    feats = np.load('train_feature_ap.npy')
    name2feat={}
    for i,name in enumerate(names):
        name2feat[name]=feats[i]
    return names,name2feat
names,name2feat = get_name2feat()
print len(names)
print len(name2feat.keys())
```

```python
# item 2 names
Item={}
for name in names:
    idx = name.split('/')[-2]
    if idx not in Item:
        Item[idx]=[]
    Item[idx].append(name)
def get_mean(item_names):
    return np.mean([name2feat[n] for n in item_names],axis=0)
# item 2 vector
vector={}
for item in Item:
    vector[item]=get_mean(Item[item])
print len(vector.keys())
```

## kmeans 测试

```python
all_items = Item.keys()
all_feats = np.array([vector[item] for item in all_items])
print len(all_items)
print all_feats.shape
```

```python
kmeans = KMeans(n_clusters=10).fit(all_feats)
```

```python
label2color=list(set(kmeans.labels_.tolist()))
label2color
```

```python
set(kmeans.labels_.tolist())
```

#### t-SNE

```python
select_items = all_items[:1000]
select_feats = np.array([vector[n] for n in select_items])
print len(select_items)
print select_feats.shape
```

```python
tsne = TSNE(n_components=2,random_state=0)
np.set_printoptions(suppress=True)
Y = tsne.fit_transform(select_feats)
```

```python
idx = kmeans.predict(vector[select_items[0]].reshape(-1,512))
idx[0]
```

## 聚类关系看着还不错

```python
plt.figure(figsize=(20,20))
plt.scatter(Y[:, 0], Y[:, 1])
for line, x, y in zip(select_items, Y[:, 0], Y[:, 1]):
    label = Item[line][0].split('/')[-3]
    label1=line[-4:]
    idx = kmeans.predict(vector[line].reshape(-1,512))[0]
    plt.annotate(label, xy=(x, y), xytext=(0, 0), textcoords='offset points',
                 color=plt.cm.Set1(label2color[idx]/10.))
```
