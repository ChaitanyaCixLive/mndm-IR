```python
import numpy as np
```

```python
names = np.load('name.npy')
labels = np.load('label.npy')
print names.shape
print labels.shape
```

```python
names[1]
```

```python
# cluster to NameSets
vector={}
for i,n in enumerate(labels):
    if str(n) not in vector:
        vector[str(n)]=[]
    vector[str(n)].append(names[i])
cluster = vector.keys()
print len(cluster)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image
base_dir = "/home/zhangxs/data/In-shop-IR/"
```

```python
cls = np.random.choice(cluster)
print 'current idx: ',cls
rst = vector[cls]
print 'num of cluster: ',len(rst)
rst[:5]
```

```python
plt.figure(figsize=(12,12))
pics = np.random.choice(vector[cls],9)
for i,p in enumerate(pics):
    plt.subplot(3,3,i+1)
    img = Image.open(base_dir+p).resize((75,75))
    plt.imshow(img)
    plt.axis('off')
```
